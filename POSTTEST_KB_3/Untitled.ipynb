```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("netflix_titles.csv")
```

```python
df
```

```python
labels = df['release_year'].value_counts().rename_axis('release_year').reset_index(name='jumlah')
labels
```

<h1>Grafik PIE</h1>

```python
plt.figure(figsize=(8,8), dpi=300)
plt.title('netflix_titles')
plt.pie(df['release_year'].value_counts(), labels=labels['release_year'], autopct='%.1f%%', startangle=90)
plt.show()
```
