# Data cleaning

```python
dataset_cl1 = df.copy()
dataset_cl2 = df2.copy()
dataset_cl3 = df3.copy()
```

```python
imputer_mean = SimpleImputer(strategy='mean')
imputer_median = SimpleImputer(strategy='median')
imputer_modus = SimpleImputer(strategy='most_frequent')
```

# modus

```python
dataset_cl2['perkSubStyle'] = imputer_modus.fit_transform(dataset_cl2[['perkSubStyle']])

dataset_cl2.isna().sum()
```

 # Mean

```python
dataset_cl2['perk5Var3'] = imputer_mean.fit_transform(dataset_cl2[['perk5Var3']])

dataset_cl2.isna().sum()
```

# K-nearest neighbour classification

```python
dataset_cl2.head()
```

```python
dataset_cl2.info()
```

```python
x = dataset_cl2[['item0','item1','item3','item4']]
y = dataset_cl2['summonerName']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0, test_size=.2)
```

```python
classifier = KNeighborsClassifier(n_neighbors = 5)
```

```python
classifier.fit(x_train,y_train)
```

```python
predicted_value = classifier.predict(x_test)
accuracy = accuracy_score(y_test, predicted_value)
```

```python
print(f'hasil prediksi: {predicted_value}',
     f'Class: {y_test}',
     f'akurasi: {accuracy}',
     sep = '\n'
)
```

```python
classifier.score(x_test,y_test)
```

# naive bayes classification

```python
x = dataset_cl2[['item0','item1','item3','item4']]
y = dataset_cl2['summonerName']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1, shuffle=False, stratify=None)
```

```python
classifier = GaussianNB()
```

```python
classifier.fit(x_train,y_train)
```

```python
predicted_value = classifier.predict(x_test)
accuracy = accuracy_score(y_test, predicted_value)
```

```python
print(f'hasil prediksi: {predicted_value}',
     f'Class: {y_test}',
     f'akurasi: {accuracy}',
     sep = '\n'
)
```

```python
classifier.score(x_test,y_test)
```
